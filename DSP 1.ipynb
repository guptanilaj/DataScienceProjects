### Title : Predict the percentage of an student based on the no. of study hours. 
### Question: What will be predicted score if a student studies for 9.25 hrs/ day?

```python
#step 1: Import libraries
import pandas
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#step 2: Importing Data
df=pandas.read_csv("http://bit.ly/w-data")
```

```python
#step 3: Seeing Data
df.head(10)
```

```python
#step 4: Inspecting data
df.describe()
```

```python
#step 4: Checking / Droping rows having null values
df.dropna(axis=0, how="any", inplace=True)
```

```python
#step 5: Visualising data
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours study and percentage scored relation graph')  
plt.xlabel('Hours studied')  
plt.ylabel('Percentage score')  
plt.show()
```

```python
#step 5: Divide the data into "attributes" (inputs) and "labels" (outputs)
x=df.iloc[:, :-1].values  
y=df.iloc[:, 1].values  
```

```python
#step 6: Split this data into training and test sets
x_train, x_test, y_train, y_test=train_test_split(x, y, train_size=0.8, random_state=0)
```

```python
#step 7: Train the algoritm
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#step 8: Testing algorithm
y_predict=regressor.predict(x_test)
#comparing acutal values with predicted values
df2=pandas.DataFrame({"predicted values": y_predict, "acutal values": y_test, "Difference": y_predict-y_test})
print(df2)
```

```python
#step 9: Evaluate the model
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predict)) 
```

```python
#step 10: Now Predict required data
z=float(input())
regressor.predict([[z]])
```
